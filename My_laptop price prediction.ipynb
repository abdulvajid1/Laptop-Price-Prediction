```python
def seprating_SSD(ssd):
    total_SSD = []
    total_num = []
    sum_SSD = 0
    if ssd == 0:
        return 0
    if '+' in ssd:
        for i in ssd.split('+'):
            if 'SSD' in i:
                total_SSD.append(i)
        for i in total_SSD:
            total_num.append(i.replace('GB','').replace('SSD','').replace('TB','000'))
        for i in total_num:
            sum_SSD = sum_SSD + eval(i)
        return sum_SSD
    else:
        return eval(ssd.replace('GB','').replace('SSD','').replace('TB','000'))
```

```python
def seprating_HDD(storage):
    total_storage = []
    total_num = []
    sum_storage = 0
    if storage == 0:
        return 0
    if '+' in storage:
        for i in storage.split('+'):
            if 'HDD' in i:
                total_storage.append(i)
        for i in total_storage:
            total_num.append(i.replace('GB','').replace('SSD','').replace('TB','000').replace('HDD','').replace('.0','')\
                             .replace('Flash Storage',''))
        for i in total_num:
            sum_storage = sum_storage + eval(i)
        return sum_storage
    else:
        return eval(storage.replace('GB','').replace('SSD','').replace('TB','000').replace('HDD','').replace('.0',''))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.tree import DecisionTreeRegressor,ExtraTreeRegressor
from sklearn.svm import LinearSVR
from sklearn.ensemble import AdaBoostRegressor,BaggingRegressor,GradientBoostingRegressor\
,HistGradientBoostingRegressor,RandomForestRegressor,StackingRegressor,VotingRegressor
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
arr = [LinearRegression,DecisionTreeRegressor,ExtraTreeRegressor,LinearSVR,\
       AdaBoostRegressor,BaggingRegressor,GradientBoostingRegressor,HistGradientBoostingRegressor,RandomForestRegressor]
```

```python
X_train.sample()
```

```python
trf1 = ColumnTransformer([
    ('ohe',OneHotEncoder(sparse_output=False,drop='first'),[0,1,2,9,10]),
    ('SS',StandardScaler(),[3,4,5,8,11,12,13,14])
                       ],remainder='passthrough')
```

```python
for i in arr:
    pipe = Pipeline([
    ('transformaton',trf1),
    ('model_apply',i())
    ])
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    print(f'for {i}')
    print('Mean ABS error= ',mean_absolute_error(y_test,y_pred))
    print('Mean SQR error= ',mean_squared_error(y_test,y_pred))
    print('R2_score = ',r2_score(y_test,y_pred))
```

```python
pipe = Pipeline([
    ('transformaton',trf1),
    ('model_apply',RandomForestRegressor(n_estimators=110,
                                         random_state=42,
                                            max_samples=0.5,
                                        max_depth=10,n_jobs=3
                                        ))])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(f'for {i}')
print('Mean ABS error= ',mean_absolute_error(y_test,y_pred))
print('Mean SQR error= ',mean_squared_error(y_test,y_pred))
print('R2_score = ',r2_score(y_test,y_pred))
```

```python
pipe = Pipeline([
    ('transformaton',trf1),
    ('model_apply',HistGradientBoostingRegressor(max_iter=405,
                                                 learning_rate=0.099999991,
                                                 max_depth=7,
                                                 max_leaf_nodes=13,
                                                 min_samples_leaf=12,max_bins=255,n_iter_no_change=100))])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(f'for {i}')
print('Mean ABS error= ',mean_absolute_error(y_test,y_pred))
print('Mean SQR error= ',mean_squared_error(y_test,y_pred))
print('R2_score = ',r2_score(y_test,y_pred))
```
